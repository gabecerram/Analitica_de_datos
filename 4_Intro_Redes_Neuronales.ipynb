1. Se procede con la carga de las librerias de trabajo

```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

2. Se procede con la grГЎfica de la funciГіn modelar

```python
x=np.linspace(-2,2,500)
yd=2*x**2+1

plt.figure()
plt.plot(x,yd, 'b', label='yd=2*x^2+1')
plt.xlabel('x')
plt.ylabel('yd')
plt.legend()
plt.grid()
plt.show()
```

3. Se procede con la implementaciГіn del modelo neuronal

```python
mnn=tf.keras.models.Sequential() #Es el esqueleto general.
mnn.add(tf.keras.layers.Dense(units=1, input_shape=[1])) # 1-NГєcleo, 1- Entrada,1- bias (Sezgo)
mnn.compile(optimizer='sgd', loss='mean_squared_error')
history=mnn.fit(x,yd, epochs=500, verbose=0)
yp=mnn.predict(x)

plt.figure()
plt.plot(x,yd, 'b', label='yd=2*x^2+1')
plt.plot(x,yp, 'r', label='y=x')
plt.grid(True)
plt.show()
```

4. Se procede con la implementaciГіn del modelo MADELIN (MultiAdaptive Linear)

```python
mnn2=tf.keras.models.Sequential() #Es el esqueleto general.
mnn2.add(tf.keras.layers.Dense(units=3, input_shape=[1], use_bias=True, activation='relu')) # 3-Regresiones, 1- Entrada,1- bias (Sezgo)
mnn2.add(tf.keras.layers.Dense(units=1, use_bias=True, activation='relu')) # 1-NГєcleo de salida, 1- bias (Sezgo)
mnn2.compile(optimizer='sgd', loss='mean_squared_error')
history=mnn2.fit(x,yd, epochs=500, verbose=1)

# Se procede al final de la grГЎfica
yp2=mnn2.predict(x)
plt.figure()
plt.plot(x,yd, 'b', label='yd=2*x^2+1')
plt.plot(x,yp2, 'r')
plt.grid(True)
plt.show()

# Para evaluar un solo valor
yp2uv=mnn2.predict(np.array([1]))
print(yp2uv)

yr=2*1^2+1
yr=3
```
