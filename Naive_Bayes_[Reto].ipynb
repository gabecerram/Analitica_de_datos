0. Se procede con la carga de las librerias de trabajo

```python
import numpy as np # Libreria por excelencia
import pandas as pd # Libreria para la comunicación de archivos en excel

# Librerias expecíficas
from sklearn.naive_bayes import GaussianNB # Libreria para el modelo Naive Bayes
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos de trabajo de la base de datos BaseAAA

```python
nxl="/content/2. LavadoActivos.xlsx"
XDB=pd.read_excel(nxl,sheet_name=0)
XDB.head(11) # Es para mostrar la base de datos cargada
print(XDB)

#Seleccionamos variables de trabajo
XD=XDB[["edad", "ingresos_usd", "gastos_usd", "monto_transado_tarjetas_usd", "porcentaje_crecimiento_patrimonio"]] #Variables de trabajo
XD.head(11)

yd=XDB[["lavado_activos"]]
yd.head(11)
```

2.  Implementamos modelo

```python
mnb=GaussianNB ()
mnb.fit(XD,yd) # Ajustar Variables entreda - salida

#Mostramos las medias de las variables
u=mnb.theta_
sigma=mnb.var_; sigma=np.sqrt(sigma)
print("edad", "ingresos_usd", "gastos_usd", "monto_transado_tarjetas_usd", "porcentaje_crecimiento_patrimonio")
print(u)
print("Las desviaciones son:")
print(sigma)
```

3. Se procede con la evaluación del modelo. Para la evaluación de este tipo de modelo se utiliza la matriz de confusión.


```python
ydp=mnb.predict(XD) # Esto es lo que el modelo aprende. yd de pronostico
cm=confusion_matrix(yd,ydp)
print(cm)

# Se determinan las metricas de la matriz de confusión
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]; TDatos=len(XDB)

# 1. Exactitud: Funcionamiento general del modelo
# Modelo que pase del 75% es muy bueno, porque de cada 4 datos que se tienen el modelo le acierta a 3 y falla en 1.
Ex=(VP+VN)/TDatos
print("Exactitud: ",Ex)

# 2. Tasa de error: % de fallos del modelo
TEr=(FP+FN)/TDatos
print("Tasa de error: ",TEr)

#3. Sensibilidad: Como se comporto frente a los positivos.
Se=VP/(VP+FN)
print("Sensibilidad: ",Se)

#4. Especificidad: Como se comporta pronosticando negativos.
Es=VN/(FP+VN)
print("Especificidad: ",Es)

#5. Precisión: Es una versión de como se comporta el modelo frente a los positivos solamente.
Pr=VP/(VP+FP)
print("Precisión: ",Pr)

#6. Predicción negativa: Como funciona el modelo pronosticando créditos prenegados
PrN=VN/(VN+FN)
print("Predicción negativa: ",PrN)
```
